```python
import pandas as pd

RATINGS_FILE = "ratings.csv"
MOVIES_FILE = "movies.csv"
TAGS_FILE = "tags.csv"

ratings = pd.read_csv(RATINGS_FILE)
movies = pd.read_csv(MOVIES_FILE)
tags = pd.read_csv(TAGS_FILE)

ratings_with_context = pd.merge(ratings, movies, on='movieId', how='left')
```
